```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,12))
p = np.linspace(0,4)
print(p)
ax.fill_between(p,5,p,
                 facecolor='green',alpha=0.5, interpolate=True, label ="zeitartig" )
ax.fill_between(p, p, 0,
                 facecolor='blue',alpha=0.5, interpolate=True, label ="raumartig" )

ax.plot(p,p,                'b-',label='$E_{\gamma}=|P|, lichtartig$ ')
ax.plot(p,1+p**2/2,         'y-',label='$E_{approx}=m+P^2/2$')
ax.plot(p,np.sqrt(p**2+1),  'k',label='$E=\sqrt{P^2+m^2}$')
ax.plot(p,p**2/2,           'r--',label='$E_{kin}=P^2/2$')
ax.plot(p,np.sqrt(p**2+1)-1,'r-',label='$E_{kin,rel}=\sqrt{P^2+m^2}-1$')

ax.axvline(1.0,0,1,c='gray',linestyle='--')
ax.axvline(2.0,0,1,c='gray',linestyle='--')

ax.set_xlabel("P [GeV]")
ax.set_ylabel("E/m")
ax.set_ylim(0,5)
ax.legend()
```
